```python
# 5 integers
numbers = [10, 20, 30, 40, 50]

# Print each element
print("The elements in the list are:")
for num in numbers:
    print(num)
```

```python
# sum of all elements
total = 0
for num in numbers:
    total = total + num

print("The sum of all elements is:", total)
```

```python
# Function to multiply two numbers
def multiply(a, b):
    return a * b
```

```python
# Function to square each number in a list
def square_list(numbers):
    squared = []
    for num in numbers:
        squared.append(num * num)
    return squared
```

```python
# Demonstrating the functions
result = multiply(5, 3)
print("Multiplication of 5 and 3 is:", result)
```

```python
my_list = [1, 2, 3, 4, 5]
print("Original list:", my_list)
print("Squared list:", square_list(my_list))
```

```python
import pandas as pd
import numpy as np

# Load the CSV file into a Pandas DataFrame
file_path = "/content/sample_data.csv"   # replace with your file path if needed
df = pd.read_csv(file_path)
```

```python
# NumPy operations on a list/array
array = np.array([10, 20, 30, 40, 50])

mean_val = np.mean(array)
sum_val = np.sum(array)
std_val = np.std(array)

print("NumPy Array Operations:")
print("Array:", array)
print("Mean:", mean_val)
print("Sum:", sum_val)
print("Standard Deviation:", std_val)
```
