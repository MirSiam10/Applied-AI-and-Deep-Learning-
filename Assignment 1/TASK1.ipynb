```python
# positive, negative, or zero
num = int(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")


```

```python
print("\nNumbers from 1 to 10:")
for i in range(1, 11):
    print(i)
```

```python
# 10 down to 1
print("\nNumbers from 10 down to 1:")
i = 10
while i >= 1:
    print(i)
    i = i - 1
```
